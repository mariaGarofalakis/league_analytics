```python
import os
from pyspark.sql import SparkSession
from delta import configure_spark_with_delta_pip

builder = (
        SparkSession.builder
        .appName("LeagueAnalytics")
        .master("local[*]")
        .config("spark.sql.extensions", "io.delta.sql.DeltaSparkSessionExtension")
        .config("spark.sql.catalog.spark_catalog", "org.apache.spark.sql.delta.catalog.DeltaCatalog")
    )

spark = configure_spark_with_delta_pip(builder).getOrCreate()
```

```python
total_standings = spark.read.format("delta").load("delta/gold/standings")
```

```python
total_standings.show(truncate=False)
```

```python
total_standings.toPandas().to_csv("output/standings_final.csv",index=False)
```

```python
total_standings_hist = spark.read.format("delta").load("delta/gold/standings_hist")
```

```python
total_standings_hist.count()
```

```python
total_standings_hist.toPandas().to_csv("output/standings_historical.csv",index=False)
```
