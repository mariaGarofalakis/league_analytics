```python
from datetime import datetime

start = datetime(2025, 8, 18, 8, 0)
end = datetime(2026, 5, 18, 8, 0)

delta = end - start
print(delta.days)       # 273
print(delta.days / 7)   # 39.0
```

```python
from datetime import datetime, timedelta

START_DATE = datetime(2025, 8, 18, 8, 0)  # Monday 08:00

def get_weekly_run_datetime(iteration: int) -> datetime:
    """
    Given an iteration number (0â€“38), return the Monday 08:00 datetime
    corresponding to that iteration.
    """
    if iteration < 0:
        raise ValueError("Iteration must be >= 0")
    
    return START_DATE + timedelta(weeks=iteration)
```

```python
print(get_weekly_run_datetime(0))
```

```python
print(get_weekly_run_datetime(1))
```

```python
print(get_weekly_run_datetime(2))
```

```python
print(get_weekly_run_datetime(39))
```

So I need to implement a for loop from 0 - 40(39)

```python
weeks = []

for week_itr in range(0,40):
    weeks.append(get_weekly_run_datetime(week_itr))

print(weeks)
```
